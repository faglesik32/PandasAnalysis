```python
file = pd.read_csv('/kaggle/input/the-office-dataset/the_office_series.csv')
file.columns
```

**TOP WRITERS**

```python
writers_rate = file[['Writers','Ratings']].groupby(['Writers']).mean()
writers_rate.sort_values(['Ratings'], ascending=False).head(5)
```

```python
writers_rate['CountEpisodes'] = file[['Writers','Ratings']].groupby(['Writers']).count()
writers_rate = writers_rate.sort_values(['Ratings'], ascending=False)
```

```python
writers_rate[writers_rate['CountEpisodes'] > 6]
```
